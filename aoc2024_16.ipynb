```python
def find_all_paths(start, goal, map):
    def dfs(current, goal, path, visited):
        if current == goal:
            paths.append(path[:])
            return
        
        x, y = current
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, down, left, right
        
        for direction in directions:
            nx, ny = x + direction[0], y + direction[1]
            if 0 <= nx < len(map) and 0 <= ny < len(map[0]) and map[nx][ny] != '#' and (nx, ny) not in visited:
                visited.add((nx, ny))
                path.append((nx, ny))
                dfs((nx, ny), goal, path, visited)
                path.pop()
                visited.remove((nx, ny))
    
    paths = []
    visited = set()
    visited.add(start)
    dfs(start, goal, [start], visited)
    return paths

with open('input_16.txt', 'r') as f:
    inputData = f.readlines()

map = []

start = None
end = None

for lineIndex in range(len(inputData)):
    line = inputData[lineIndex].strip()
    lineMap = []
    for columnIndex in range(len(line)):
        lineMap.append(line[columnIndex])
        if line[columnIndex] == 'S':
            start = (lineIndex, columnIndex)
        if line[columnIndex] == 'E':
            end = (lineIndex, columnIndex)
    map.append(lineMap)

# print(start, end)
# print(map)



all_paths = find_all_paths(start, end, map)

print(len(all_paths))
print(all_paths)


total = 0
for sequence in sequences:
    totalSequence = 0
    currentStep = 'R'
    for step in sequence:
        if step != currentStep:
            totalSequence += 1000
            currentStep = step
        totalSequence += 1
    if totalSequence < total or total == 0:
        total = totalSequence

print(total)
```

```python
import heapq

def has_changed_direction(current, neighbor, current_direction):
    direction = (neighbor[0] - current[0], neighbor[1] - current[1])
    return direction != current_direction

def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def find_lowest_cost_path(start, goal, map):
    # Priority queue
    open_set = []
    heapq.heappush(open_set, (0, start, None))  # (cost, position, direction)

    # Dictionaries to keep track of the path
    came_from = {}
    g_score = {start: 0}
    f_score = {start: heuristic(start, goal)}

    while open_set:
        current_cost, current, current_direction = heapq.heappop(open_set)

        if current == goal:
            # Reconstruct path
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            path.reverse()
            return path, current_cost

        for neighbor in get_neighbors(current, map):
            tentative_g_score = g_score[current] + 1  # Cost of moving to the neighbor
            if has_changed_direction(current, neighbor, current_direction):
                tentative_g_score += 1000  # Additional cost for changing direction

            if neighbor not in g_score or tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = g_score[neighbor] + heuristic(neighbor, goal)
                heapq.heappush(open_set, (f_score[neighbor], neighbor, (neighbor[0] - current[0], neighbor[1] - current[1])))

    return None, float('inf')  # No path found

def get_neighbors(position, map):
    neighbors = []
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, down, left, right
    for direction in directions:
        neighbor = (position[0] + direction[0], position[1] + direction[1])
        if 0 <= neighbor[0] < len(map) and 0 <= neighbor[1] < len(map[0]) and map[neighbor[0]][neighbor[1]] != '#':
            neighbors.append(neighbor)
    return neighbors

with open('input_16.txt', 'r') as f:
    inputData = f.readlines()

map = []

start = None
end = None

for lineIndex in range(len(inputData)):
    line = inputData[lineIndex].strip()
    lineMap = []
    for columnIndex in range(len(line)):
        lineMap.append(line[columnIndex])
        if line[columnIndex] == 'S':
            start = (lineIndex, columnIndex)
        if line[columnIndex] == 'E':
            end = (lineIndex, columnIndex)
    map.append(lineMap)

print(start, end)
# print(map)

startingDirection = (0,1)

path, currentCost = find_lowest_cost_path(start, end, map)

print(path)

total = 0
previousCoord = start
currentDirection = startingDirection

print('calculating total')
print('currentCost', currentCost)

for coord in path:
    if coord == start:
        previousCoord = coord
        continue
    if has_changed_direction(previousCoord, coord, currentDirection):
        total += 1000
        currentDirection = (coord[0] - previousCoord[0], coord[1] - previousCoord[1])
    total += 1
    previousCoord = coord

print(total)
```

```python
import heapq
import math
from collections import defaultdict


def parse_input():
    with open("input_16.txt", "r") as file:
        data = file.read()
    grid = data.splitlines()
    n, m = len(grid), len(grid[0])
    start = end = None
    for i, row in enumerate(grid):
        for j, v in enumerate(row):
            if v == "S":
                start = (i, j)
            if v == "E":
                end = (i, j)
    return grid, n, m, start, end


GRID, N, M, START, END = parse_input()

print(N, M, START, END)

def solve(pt2=False):
    scores = defaultdict(lambda: math.inf)
    i, j = START
    di, dj = 0, 1
    pq = [(0, i, j, di, dj, {(i, j)})]
    res = math.inf
    all_tiles = set()
    while pq:
        score, i, j, di, dj, tiles = heapq.heappop(pq)
        if (i, j) == END:
            if not pt2:
                return score
            res = min(res, score)
            if score == res:
                all_tiles |= tiles
        for nscore, ni, nj, ndi, ndj in (
            (score + 1, i + di, j + dj, di, dj),
            (score + 1001, i - dj, j + di, -dj, di),
            (score + 1001, i + dj, j - di, dj, -di),
        ):
            if (
                0 <= ni < N
                and 0 <= nj < M
                and GRID[ni][nj] != "#"
                and (
                    scores[ni, nj, ndi, ndj] >= nscore
                    if pt2
                    else scores[ni, nj, ndi, ndj] > nscore
                )
            ):
                scores[ni, nj, ndi, ndj] = nscore
                heapq.heappush(pq, (nscore, ni, nj, ndi, ndj, tiles | {(ni, nj)}))
    return len(all_tiles)


def part_one():
    return solve()


def part_two():
    return solve(True)


print(f"Part 1: {part_one()}")
print(f"Part 2: {part_two()}")
```

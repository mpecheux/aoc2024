```python
def levelsAreAllIncraesing(levels):
    for i in range(1, len(levels)):
        if (levels[i] <= levels[i - 1]):
            return False
        if (levels[i]-levels[i-1]>3):
            return False
    return True

def levelsAreAllDecreasing(levels):
    for i in range(1, len(levels)):
        if (levels[i] >= levels[i - 1]):
            return False
        if (levels[i-1]-levels[i]>3):
            return False
    return True

datas = []

total = 0

with open('input_2.txt', 'r') as f:
    lines = f.readlines()



    for line in lines:
        levelsStr = line.split(' ')
        
        levels = [int(level) for level in levelsStr]

        if (levels[0] == levels[1]):
            datas.append(levels)
            continue
        elif (levels[0] < levels[1]):
            if levelsAreAllIncraesing(levels):
                total += 1
            else:
                datas.append(levels)
        else:
            if levelsAreAllDecreasing(levels):
                total += 1
            else:
                datas.append(levels)

leftDatas = datas.copy()

for data in leftDatas:
    # print(data)
    for i in range(0, len(data)):
        #remove element in levels
        levels= data.copy()
        levels.pop(i)
        # print(levels)
        if (levels[0] == levels[1]):
            continue
        elif (levels[0] < levels[1]):
            if levelsAreAllIncraesing(levels):
                total += 1
                datas.remove(data)
                break
        else:
            if levelsAreAllDecreasing(levels):
                total += 1
                datas.remove(data)
                break

print('leftDatas')
print(datas)
print(total)
        


    
```

```python
def isOnlyZeros(stone):
    for char in stone:
        if char != '0':
            return False
    return True

def deleteLeadingZeros(half):   
    while half[0] == '0' and not isOnlyZeros(half):
        half = half[1:]
    
    if isOnlyZeros(half):
        half = '0'
    
    return half

def splitInHalf(stone):
    half = len(stone)//2
    firstHalf = stone[:half]
    secondHalf = stone[half:]
    secondHalf = deleteLeadingZeros(secondHalf)
    return firstHalf, secondHalf

def getTotal(cacheStones):
    total=0
    for stone, count in cacheStones.items():
        total+=count
    return str(total)

with open('input_11.txt', 'r') as f:
    inputData = f.readlines()

stones = []
for lineIndex in range(len(inputData)):
    line = inputData[lineIndex].strip()
    stones = line.split(' ')


# stones = ['125','17']
print(stones)

cacheStones = {}

for stone in stones:
    cacheStones[stone] = 1

print(cacheStones)

i=0
addingTotal = 0
while i!=75:
    newCacheStones = {}

    for stone, count in cacheStones.items():
        if stone == '0':
            newStone = '1'
            if newStone in newCacheStones:
                newCacheStones[newStone]=newCacheStones[newStone] + count
            else:
                newCacheStones[newStone]=count
        elif len(stone)%2 == 0:
            firstHalf, secondHalf = splitInHalf(stone)
            newStone1 = firstHalf
            newStone2 = secondHalf
            if newStone1 in newCacheStones:
                newCacheStones[newStone1] =  newCacheStones[newStone1] + count
            else:
                newCacheStones[newStone1] = count

            if newStone2 in newCacheStones:
                newCacheStones[newStone2] =  newCacheStones[newStone2] + count
            else:
                newCacheStones[newStone2] = count
        else:
            newStone = int(stone)*2024
            if newStone in newCacheStones:
                newCacheStones[newStone] =  newCacheStones[newStone] + count
            else:
                newCacheStones[str(newStone)] = count
    cacheStones = newCacheStones
    # print(str(i+1) + ' blink ' + getTotal(cacheStones) + ' cacheStones ' + str(cacheStones))
    i+=1


total=0
for stone, count in cacheStones.items():
    total+=count

print('cacheStones ' + str(len(cacheStones)))
print('total ' + str(total))
```

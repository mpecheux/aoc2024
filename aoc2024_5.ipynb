```python
with open('input_5.txt', 'r') as f:
    inputData = f.readlines()

allowedAfter = {}
allowedBefore = {}
prints=[]

for line in inputData:
    line = line.strip()
    if '|' in line:
        key = line.split('|')[0]
        value = line.split('|')[1]
        if key in allowedAfter:
            allowedAfter[key].append(value)
        else:
            allowedAfter[key] = [value]
        if value in allowedBefore:
            allowedBefore[value].append(key)
        else:
            allowedBefore[value] = [key]
    elif ',' in line:
        prints.append(line)

validPrints = []
invalidPrints = []

for printLine in prints:
    printLine = printLine.split(',')
    valid = True
    for i in range(len(printLine)-1):
        for j in range(i+1, len(printLine)):
            if printLine[j] in allowedBefore[printLine[i]]:
                valid = False
                break
        if printLine[i+1] not in allowedAfter[printLine[i]]:
            valid = False
            break
    if valid:
        validPrints.append(printLine)
    else:
        invalidPrints.append(printLine)

# print(len(validPrints))

total = 0
for line in validPrints:
    middleIndex = (len(line)-1) // 2
    total += int(line[middleIndex])

print('total part 1 : ' + str(total))

correctedPrints = []

for line in invalidPrints:
    print('invalidPrints ')
    print(line)
    isNotValid = True
    # print('line is not valid ' + str(line))
    while isNotValid:
        isNotValid = False
        for i in range(len(line)-1):
            if isNotValid:
                break
            for j in range(i+1, len(line)):
                if line[j] in allowedBefore[line[i]]:
                    #switch place 
                    a = line[i]
                    b = line[j]
                    line[i] = b
                    line[j] = a
                    isNotValid = True
                    break
            if line[i+1] not in allowedAfter[line[i]]:
                a = line[i]
                b = line[i+1]
                line[i] = b
                line[i+1] = a
                isNotValid = True
                break
    correctedPrints.append(line)
    # print('line is valid ' + str(line))
        


correctedTotal = 0
for line in correctedPrints:
    middleIndex = (len(line)-1) // 2
    correctedTotal += int(line[middleIndex])
    print(correctedTotal)

print('corrected prints')
print(len(correctedPrints))
print('corrected score')
print(correctedTotal)


```

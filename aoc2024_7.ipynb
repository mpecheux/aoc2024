```python
# Part 1
import copy
import re
import operator

def concatenate(a, b):
    return int(str(a) + str(b))

def evaluate_left_to_right(expression, targetResult):
    # Define operator functions
    operators = {
        '+': operator.add,
        '*': operator.mul,
        '|': concatenate
    }
    
    # Split the expression into numbers and operators
    tokens = re.findall(r'\d+|\+|\-|\*|\/|\|', expression)
    
    # Initialize the result with the first number
    result = int(tokens[0])
    
    # Iterate through the tokens and apply each operation from left to right
    i = 1
    while i < len(tokens):
        op = tokens[i]
        num = int(tokens[i + 1])
        result = operators[op](result, num)
        if (result > targetResult):
            return -1
        i += 2
    
    return result

def isValidEquation(result, numbers):
    operators = []
    for i in range(len(numbers)-1):
        if not operators:
            operators.append(['+'])
            operators.append(['*'])
        else:
            operatorsCopy = copy.deepcopy(operators)
            for operator in operatorsCopy:
                operator.append('+')
            for operator in operators:
                operator.append('*')
            operators += operatorsCopy

    for operator in operators:
        stringOperation = ''
        i = 0
        for number in numbers:
            if (i == len(numbers)-1):
                strNumber = str(number)
                stringOperation = stringOperation + strNumber
            else:
                strNumber = str(number)
                strOperator = operator[i]
                stringOperation = stringOperation + strNumber + strOperator
            i += 1

        operationResult=evaluate_left_to_right(stringOperation, int(result))
  
        if int(operationResult) == int(result):
            return True

    return False

with open('input_7.txt', 'r') as f:
    inputData = f.readlines()

total = 0

print('Nombre d\'openations:', len(inputData))

unValidOperation = []

for line in inputData:
    line = line.strip()
    result = line.split(': ')[0]
    numbers = line.split(': ')[1].split(' ')
    numbers = [int(num) for num in numbers]
    if isValidEquation(result, numbers):
        total += int(result)
    else:
        unValidOperation.append(line)
    i += 1

print('Nombre d\'openations non valides:', len(unValidOperation))

print('total:', total)
```

```python
import copy

    
def checkDesign(design, towels):
    if (design == ''):
        return True
   
    for towel in towels:
        if (design.endswith(towel)):
            if checkDesign(design[:len(design)-len(towel)], towels):
                return True
 
    return False

def checkDesignAndCountValidSolution(design, towels, cache):        

    if (design == ''):        
        return 1

    count = 0
    for towel in towels:
        if (design.endswith(towel)):
            subDesign = design[:len(design)-len(towel)]
            if (subDesign in cache):
                count += cache[subDesign]
            else:
                countSub = checkDesignAndCountValidSolution(subDesign, towels, cache)
                count = count + countSub
                if (subDesign != ''):
                    cache[subDesign] = countSub
                
    return count

with open('input_19.txt', 'r') as f:
    inputData = f.readlines()


towels = set()
designs = set()


for lineIndex in range(len(inputData)):
    line = inputData[lineIndex].strip()
    if (lineIndex == 0):
        towels = line.split(', ')
    else:
        if (line.strip() != ''):
            designs.add(line)


print(towels)
print(designs)

total = 0

goodDesigns = set()
cache = {}

for design in designs:    
    if checkDesign(design, towels):        
        goodDesigns.add(design)
        total += 1    

countTotal = 0

# goodDesigns = set()
# goodDesigns.add('bw')


for design in goodDesigns:
    # print(design)
    count = checkDesignAndCountValidSolution(design, towels, cache)
    # print('count for design: ', count)
    countTotal += count
    
print(cache)    

print(countTotal)
```

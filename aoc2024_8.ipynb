```python
#Part 1

def calculateNodeCoordinate(coord1, coord2):
    return (coord1[0] - (coord2[0]-coord1[0])), (coord1[1] - (coord2[1]-coord1[1]))

def isInGrid(coord, gridSize):
    # print(coord)
    isInGrid = coord[0] >= 0 and coord[0] < gridSize[0] and coord[1] >= 0 and coord[1] < gridSize[1]
    return isInGrid

with open('input_8.txt', 'r') as f:
    inputData = f.readlines()

letterCoordinates = {}

gridSize = (len(inputData), len(inputData[0].strip()))

print(gridSize)

for i in range(len(inputData)):
    line = inputData[i].strip()
    for j in range(len(line)):
        if line[j] != '.':
            letter  = line[j]
            if letter not in letterCoordinates:
                letterCoordinates[letter] = []
            letterCoordinates[letter].append((i,j))
            
# print(letterCoordinates)

nodeCoordinates = []

for letter in letterCoordinates:
    for i in range(len(letterCoordinates[letter])):
        for j in range(len(letterCoordinates[letter])):
            if letterCoordinates[letter][i] != letterCoordinates[letter][j]:
                nodeCorrdinate = calculateNodeCoordinate(letterCoordinates[letter][i], letterCoordinates[letter][j])
                if nodeCorrdinate not in nodeCoordinates and isInGrid(nodeCorrdinate, gridSize):
                    nodeCoordinates.append(nodeCorrdinate)

print('results ' + str(len(nodeCoordinates)))
                
        
```

```python
#Part 2

def calculateNodeCoordinates(coord1, coord2):
    pas0 = (coord2[0]-coord1[0])
    pas1 = (coord2[1]-coord1[1])

    inGrid = True
    nodeCoords = []
    i = 0
    while inGrid:
        nodeCoord = (coord1[0] - (i*pas0), coord1[1] - (i*pas1))
        if isInGrid(nodeCoord, gridSize):
            nodeCoords.append(nodeCoord)
        else:
            inGrid = False
        i += 1

    return nodeCoords

def isInGrid(coord, gridSize):
    # print(coord)
    isInGrid = coord[0] >= 0 and coord[0] < gridSize[0] and coord[1] >= 0 and coord[1] < gridSize[1]
    return isInGrid

with open('input_8.txt', 'r') as f:
    inputData = f.readlines()

letterCoordinates = {}

gridSize = (len(inputData), len(inputData[0].strip()))

print(gridSize)

for i in range(len(inputData)):
    line = inputData[i].strip()
    for j in range(len(line)):
        if line[j] != '.':
            letter  = line[j]
            if letter not in letterCoordinates:
                letterCoordinates[letter] = []
            letterCoordinates[letter].append((i,j))
            
print(letterCoordinates)

nodeCoordinates = []

for letter in letterCoordinates:
    for i in range(len(letterCoordinates[letter])):
        for j in range(len(letterCoordinates[letter])):
            if letterCoordinates[letter][i] != letterCoordinates[letter][j]:
                newNodeCorrdinates = calculateNodeCoordinates(letterCoordinates[letter][i], letterCoordinates[letter][j])
                # print(newNodeCorrdinates)
                for coord in newNodeCorrdinates:
                    if coord not in nodeCoordinates:
                        nodeCoordinates.append(coord)


print('results ' + str(len(nodeCoordinates)))
                
#plot the results
import matplotlib.pyplot as plt

y = [(11-coord[1]) for coord in nodeCoordinates]
x = [coord[0] for coord in nodeCoordinates]

plt.scatter(x, y)
```

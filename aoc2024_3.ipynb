```python
import re

with open('input_3.txt', 'r') as f:
    inputData = f.readlines()

# inputDataEx = "xmul(2,4)%&mul[3,7]!@^do_not_mul(5,5)+mul(32,64]then(mul(11,8)mul(8,5))"

pattern = r"mul\(\d+,\d+\)"
total = 0

for line in inputData:
    matches = re.findall(pattern, line)
    # matches = re.findall(pattern, inputDataEx)
    print(matches)

    for match in matches:
        match = match.replace('mul(', '')
        match = match.replace(')', '')
        a, b = match.split(',')
        product = int(a) * int(b)
        # print(f'{a} * {b} = {product}')
        total = total + product

print(total)
```

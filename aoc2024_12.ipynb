```python
import copy

def isValid(coord, map):
    if coord[0] < 0 or coord[0] >= len(map):
        return False
    if coord[1] < 0 or coord[1] >= len(map[0]):
        return False
    return True

def addIfNeightbours(newCoords, coord, area):
    upCoord = (coord[0]-1, coord[1])
    downCoord = (coord[0]+1, coord[1])
    leftCoord = (coord[0], coord[1]-1)
    rightCoord = (coord[0], coord[1]+1)

    if upCoord not in area:
        newCoords.append(upCoord)
    if downCoord not in area:
        newCoords.append(downCoord)
    if leftCoord not in area:
        newCoords.append(leftCoord)
    if rightCoord not in area:
        newCoords.append(rightCoord)

def findAreaForLetter(map, currentLetter, currentCoord, selectedCoords):
    area = []

    selectedCoords.append(currentCoord)
    area.append(currentCoord)

    newCoords = []
    newCoords.append((currentCoord[0]-1, currentCoord[1]))
    newCoords.append((currentCoord[0]+1, currentCoord[1]))
    newCoords.append((currentCoord[0], currentCoord[1]-1))
    newCoords.append((currentCoord[0], currentCoord[1]+1))

    while newCoords:
        currentCoords = copy.deepcopy(newCoords)
        for coord in currentCoords:

            if coord not in selectedCoords and isValid(coord, map) and map[coord[0]][coord[1]] == currentLetter:
                area.append(coord)
                selectedCoords.append(coord)
                addIfNeightbours(newCoords, coord, area)
            newCoords.remove(coord)

    return area

def areAlignedNeighbors(area, coord):
    leftCoord = (coord[0], coord[1]-1)
    rightCoord = (coord[0], coord[1]+1)
    upCoord = (coord[0]-1, coord[1])
    downCoord = (coord[0]+1, coord[1])
    if leftCoord in area and rightCoord in area:
        return True
    if upCoord in area and downCoord in area:
        return True
    return False

def keepEdgeCoords(area):
    edgeCoords = []
    for coord in area:
        leftCoord = (coord[0], coord[1]-1)
        rightCoord = (coord[0], coord[1]+1)
        upCoord = (coord[0]-1, coord[1])
        downCoord = (coord[0]+1, coord[1])
        if leftCoord not in area or rightCoord not in area or upCoord not in area or downCoord not in area:
            edgeCoords.append(coord)
    return edgeCoords

def countContinousList(list):
    count = 1
    for i in range(len(list)-1):
        if list[i+1] != list[i]+1:
            count = count + 1
    return count

def hasNeighbor(area, coord):
    if coord in area:
        return True
    return False

def calculateSides(area):
    if len(area) == 1:
        return 4
    fullArea = copy.deepcopy(area)
    area = keepEdgeCoords(area)

    topLineSides = {}
    bottomLineSides = {}
    leftColumnSides = {}
    rightColumnSides = {}

    for coord in area:
        line = coord[0]
        column = coord[1]
        if not hasNeighbor(fullArea, (line-1, column)):
            if line in topLineSides:
                if not column in topLineSides[line]:
                    topLineSides[line].append(column)
            else:
                topLineSides[line] = [column]
        if not hasNeighbor(fullArea, (line+1, column)):
            if line in bottomLineSides:
                if not column in bottomLineSides[line]:
                    bottomLineSides[line].append(column)
            else:
                bottomLineSides[line] = [column]
        if not hasNeighbor(fullArea, (line, column-1)):
            if column in leftColumnSides:
                if not line in leftColumnSides[column]:
                    leftColumnSides[column].append(line)
            else:
                leftColumnSides[column] = [line]
        if not hasNeighbor(fullArea, (line, column+1)):
            if column in rightColumnSides:
                if not line in rightColumnSides[column]:
                    rightColumnSides[column].append(line)
            else:
                rightColumnSides[column] = [line]

    
    for topLineSide in topLineSides.values():
        topLineSide.sort()
    for bottomLineSide in bottomLineSides.values():
        bottomLineSide.sort()
    for leftColumnSide in leftColumnSides.values():
        leftColumnSide.sort()
    for rightColumnSide in rightColumnSides.values():
        rightColumnSide.sort()
    
    
    print(topLineSides)
    print(bottomLineSides)
    print(leftColumnSides)    
    print(rightColumnSides)

    side = 0
    for topLineSide in topLineSides.values():
        side = side + countContinousList(topLineSide)
    for bottomLineSide in bottomLineSides.values():
        side = side + countContinousList(bottomLineSide)
    for leftColumnSide in leftColumnSides.values():
        side = side + countContinousList(leftColumnSide)
    for rightColumnSide in rightColumnSides.values():
        side = side + countContinousList(rightColumnSide)


    return side


def calculateSidesOld(area):

    side = 0

    for coord in area:
        neighbor = 0
        leftCoord = (coord[0], coord[1]-1)
        rightCoord = (coord[0], coord[1]+1)
        upCoord = (coord[0]-1, coord[1])
        downCoord = (coord[0]+1, coord[1])
        if leftCoord in area:
            neighbor = neighbor + 1
        if rightCoord in area:
            neighbor = neighbor + 1
        if upCoord in area:
            neighbor = neighbor + 1
        if downCoord in area:
            neighbor = neighbor + 1
        if neighbor == 0:
            side = 4
        elif neighbor == 1:
            print('oneNeighbor')
            side = side + 2
        elif neighbor == 2:
            print('twoNeighbor')
            if areAlignedNeighbors(area, coord):
                print('alignedNeighbors')
                side = side
            else:
                side = side + 2

    return side

def calculAreaCost(area):
    
    totaSize = calculateSides(area)
    print(totaSize)
    total = totaSize*len(area)
    
    return total

with open('input_12.txt', 'r') as f:
    inputData = f.readlines()

map = []

for lineIndex in range(len(inputData)):
    line = inputData[lineIndex].strip()
    lineMap = []
    for columnIndex in range(len(line)):
        lineMap.append(line[columnIndex])
    map.append(lineMap)

# print(str(len(map)))
# print(str(len(map[0])))

selectedCoords = []
areas = []

currentLetter = ''

for lineIndex in range(len(map)):
    for columnIndex in range(len(map[lineIndex])):
        currentCoord = (lineIndex, columnIndex)

    

        if currentCoord not in selectedCoords:
            currentLetter = map[lineIndex][columnIndex]
            area = findAreaForLetter(map, currentLetter, currentCoord, selectedCoords)
            areas.append(area)
total = 0
for area in areas:
    print(map[area[0][0]][area[0][1]])
    total = total + calculAreaCost(area)

print('total cost ' + str(total))

```
